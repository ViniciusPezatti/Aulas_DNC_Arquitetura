```python
import pandas as pd
from faker import Faker
import random

# Inicializando Faker
fake = Faker()

# Função para gerar dados fictícios
def generate_data(n):
    data = []  # Corrigido: inicializando a lista
    for _ in range(n):
        name = fake.name()
        age = random.randint(18, 65)
        city = fake.city()
        balance = round(random.uniform(1000, 1000000), 2)
        account_number = fake.bban()

        # Adiciona os dados como uma tupla na lista
        data.append((name, age, city, balance, account_number))

    return data

# Gerar dados para 50.000 registros
dados = generate_data(50000)

# Criar um DataFrame com Pandas
df = pd.DataFrame(dados, columns=['Name', 'Age', 'City', 'Balance', 'Account_Number'])

# Exibir as primeiras linhas do DataFrame
df.head()
```

```python
# Definir a idade para o filtro
idade = 60

# Filtrar clientes com idade maior que a especificada
df_idade_acima = df[df['Age'] > idade]

# Exibir os 10 primeiros clientes com idade acima do limite
print(df_idade_acima.head(10))

print('-----' * 50)

# Contar o número total de clientes acima da idade especificada
print('Clientes acima de', idade, "=", len(df_idade_acima), 'clientes')
```

```python
# Definir os limites de idade e condição para saldo
idade_min = 30
idade_max = 50

# Filtrar clientes com idade entre 30 e 50 anos e idade da conta maior ou igual a 10 anos
df_clientes_filtrados = df.loc[
    (df['Age'] >= idade_min) &
    (df['Age'] <= idade_max)
]

# Ordenar pelo saldo ('Balance') em ordem decrescente
df_clientes_filtrados = df_clientes_filtrados.sort_values(by='Balance', ascending=False)

# Exibir os 10 primeiros clientes com maior saldo dentro da faixa etária
print(df_clientes_filtrados.head(10))

# Exibir o número total de clientes nessa condição
print('-----' * 50)
print()
print('Total de clientes entre', idade_min, 'e', idade_max, 'anos =', len(df_clientes_filtrados), 'clientes')
```
